11)print the total number of duplicate rows

```python
#rows containting duplicate data
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
#drop the dupicate
car=car.drop_duplicates()
car.head()
```

13)Drop the missing  values from the dataset 

```python
#find the null values
print(car.isnull().sum())
```

```python
#drop the missing value
car=car.dropna()
car.count()
```

14)plot a histogram to find the number of cars per brand


```python
car.hp.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("number of cars by hp")
plt.ylabel("number of cars")
plt.xlabel("hp");
```
